```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

# Data collecting and procesing

```python
breast_Cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_Cancer_dataset)
```

```python
dataframe = pd.DataFrame(breast_Cancer_dataset.data, columns = breast_Cancer_dataset.feature_names)
```

```python
dataframe.head()
```

```python
dataframe["Label"] = breast_Cancer_dataset.target
```

```python
dataframe.head()
```

```python
dataframe.shape
```

```python
dataframe.info()
```

```python
dataframe.describe()
```

```python
dataframe['Label'].value_counts()
```

```python
x = dataframe.drop(columns='Label',axis=1)
y = dataframe['Label']
```

```python
x
```

```python
y
```

# Splitting the data into train and test

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size = 0.2, random_state=2)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

# standardlize the data 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
X_train_std
```

# Building the NN model


```python
import tensorflow as tf
```

```python
print('hellp')
```

```python
# seting the random state
tf.random.set_seed(42)

# making the model 
model_1 = tf.keras.Sequential([tf.keras.layers.Flatten(input_shape=(30,)),
                               tf.keras.layers.Dense(20,activation="relu"),
                               tf.keras.layers.Dense(2,activation="sigmoid")])

# compiling the model
model_1.compile(loss = tf.keras.losses.SparseCategoricalCrossentropy(),
                optimizer = tf.keras.optimizers.Adam(),
                metrics = ["accuracy"])

# fiting the model
history_1 = model_1.fit(tf.expand_dims(X_train_std,axis=-1),Y_train,epochs=10)
```

```python
pd.DataFrame(history_1.history).plot()
```

```python
Y_pred = model_1.predict(X_test_std)
Y_pred[0]
```
